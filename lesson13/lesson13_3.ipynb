```python
#標準寫法
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.axes import Axes

# 1.製作figure
fig:Figure = plt.figure()

# 2.在1.的figure上製作一個axes
#ax_1是變數名,乘載的資料類型為Axes
ax_1:Axes = fig.add_subplot(1, 1, 1)
ax_1.set_title('this is a chart')

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y, color='g', marker='<', linestyle='--', linewidth=2, label='數據1', alpha=0.8, markersize=10)


# 4.顯示圖
plt.show()
```

```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure() 
#fig是變數名,plt是前面導入的pyplot這個子模組的變數, 
#figure是pyplot裡面的一個函數,用來新建一張空白的畫布

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```
