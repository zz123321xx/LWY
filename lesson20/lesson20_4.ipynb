```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips.columns = ['總票價','小費','吸煙者','日期','時間','大小']
tips['小費比例'] = tips['小費'] / tips['總票價'] * 100
tips

```

```python
#折分為多個步驟,執行groupby
grouped = tips.groupby(['日期','吸煙者'])
grouped_pct = grouped[['小費','總票價']]
grouped_pct.agg([('平均','mean'),('標準差','std')])
```

```python
#拆分為多個步驟,執行groupby
def peak_to_peak(arr):
    return arr.max() - arr.min()
    
grouped = tips.groupby(['日期','吸煙者'])
grouped_pct = grouped['小費']
grouped_pct.agg([('平均','mean'),('標準差','std'),('最多','max'),('最少','min'),("差值",peak_to_peak)])
```
