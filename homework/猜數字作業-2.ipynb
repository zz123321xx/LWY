```python
import random

# 設定原始最大值與最小值
ori_max = 10
ori_min = 1

# 初始化最大值與最小值
max = ori_max
min = ori_min

# 遊戲次數與猜測次數初始化
games = 1
times = 0

# 隨機產生答案
ans = random.randint(min, max)

# 是否繼續遊戲的變數
yn = "y"

print("==========遊戲開始==========")

while True:
    # 判斷是否繼續遊戲
    if yn == "y" or yn == "Y":
        try:
            # 取得玩家輸入
            player_ans = int(input(f"請輸入你猜的數字,數字為{min}到{max}"))
            # 猜對答案
            if player_ans == ans:
                times += 1
                print(f"沒錯!答案是{ans}!你好蔥茗鴨!")
                print(f"你本次遊戲總共猜了{times}次")
                # 詢問是否再玩一局
                while True:
                    yn = input("是否再來一局")
                    if yn == "y" or yn == "Y":
                        # 重設遊戲參數
                        max = ori_max
                        min = ori_min
                        ans = random.randint(min, max)
                        times = 0
                        games += 1
                        print("==========遊戲再開==========")
                        break
                    elif yn == "n" or yn == "N":
                        print("==========遊戲結束==========")
                        break
                    else:
                        print("輸入錯誤的指令，請輸入y或n")
            # 猜的數字太小
            elif player_ans < ans:
                if player_ans < min:
                    print("太小了!比你要猜的數字還小!")
                else:
                    print("再大一點")
                    times += 1
                    min = player_ans + 1
                    continue
            # 猜的數字太大
            else:
                if player_ans > max:
                    print("太大了!比你要猜的數字還大!")
                else:
                    print("再小一點")
                    times += 1
                    max = player_ans - 1
                    continue
        except ValueError:
            # 處理輸入非整數的錯誤
            print("格式錯誤")
        except Exception as e:
            # 處理其他例外
            print(e)
    else:
        # 遊戲結束，顯示總遊玩次數
        print("感謝你的遊玩")
        print(f"你總共遊玩了{games}次!")
        break
```
